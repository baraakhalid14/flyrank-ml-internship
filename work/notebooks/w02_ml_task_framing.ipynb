## 3. Success metric

*One metric you can defend. What number means 'good'?*

```python
"""
Success metric: NDCG (Normalized Discounted Cumulative Gain)

NDCG measures how well the model ranks the most relevant search results near the top. A higher NDCG score means the search results are ordered more effectively for the user.
"""
```

## 5. Why ML beats a fixed rule here

```python
"""
Machine learning is better than a fixed rule because search relevance depends on many factors, such as the user's query, item features, and past user interactions. A fixed rule cannot capture these complex patterns, while a machine learning model can learn from data and improve the ranking of search results over time.
"""
```
